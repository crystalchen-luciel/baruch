```python
# The introduction function prints an introduction message explaining the program. It does not take any parameters. It only prints once at the start of the program.
def introduction():
    print("This program calculates the average of three integers.\nAfter calculating the average of the three integers, it compares the original three integers to the calculated average of the three integers.\nIt will then tell you how many of the original integers are equal to the average.\nThe program will continue to run until you enter 0 three times.\nWhen you enter 0 three times, the program will end and print how many sets of three data values were entered and processed.\nCreated by: Crystal Chen.\n")

# This function will calculate the average and send the answer computed back to the main function
# The parameters are the three numbers
# This function returns the average
def findaverage(num1, num2, num3):
  average = float((num1 + num2 + num3) / 3)
  return average

# This function compares each input (the three integers) to the average
# It will tell the user if the integer is less than, greater than, or equal to the average
# It also counts the amount of integers that is equal to the average
# The parameters are the average (average of the three numbers), num1 (the first input number), num2 (the second input number), and num3 (the third input)
# This function does not return anything
def comparetoavg(average, num1, num2, num3):
  countequalavg = 0

  for num in [num1, num2, num3]:
        if num == average:
            print(f"{num} is equal to the average.")
            countequalavg += 1
        elif num < average:
            print(f"{num} is less than the average.")
        else:
            print(f"{num} is greater than the average.")

  print(f"The number of original values equal to the average is: {countequalavg}.\n\n\n")

# The main function is the main function of the program. It takes three integer inputs from the user and calculates the average by sending the numbers to the findaverage function
# After it receives the average, it passes information about the three integers and the average to the comparetoavg function and increments 1 to the sets_processed count
# This loop continues until three zeros are entered by the user
# Once three zeros is entered by the user, the program will print how many sets it has processed and terminate
def main():
  sets_processed = 0
  introduction()

  while True:
    num1 = int(input("Enter an integer value: "))
    num2 = int(input("Enter an integer value: "))
    num3 = int(input("Enter an integer value: "))
    print(f"The three numbers you entered (perspectively) are: {num1}, {num2}, {num3}.\n")
    
    if num1 == num2 == num3 == 0:
      print(f"This program has processed {sets_processed} sets of data.")
      break

    average = findaverage(num1, num2, num3)
    print(f"The average of {num1}, {num2}, {num3}, is {average:.3f}.")

    sets_processed += 1
    comparetoavg(average, num1, num2, num3)

# Call the main function to execute the code
main()
```
