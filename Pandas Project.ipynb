```python
df
```

```python
# Assigning the columns new meaninful names

new_column_names = {
    'Boro': 'Borough',
    'LifeCycle': 'BuildingStatus',
    'WorkTypeGeneral': 'WorkType',
    'OMOStatusReason': 'BuildingClosingReason',
    'OMOAwardAmount': 'AmountAwarded',
    'OMOAwardDate': 'AmountAwardedDate',
    'OMOCreateDate': 'ReportCreationDate',
    'IsCommercialDemolition': 'CommercialDemolition',
    'ServiceChargeFlag': 'ServiceCharge',
    'OMODescription': 'BuildingClosureDescription'
}

# Renaming multiple columns
df.rename(columns=new_column_names, inplace=True)
```

```python
# Dropping the rows containing NaN values, making it easier to analyze the data
# Did not replace missing values with the mean as I believe it would skew the data too much
# There are still 730 rows after removing all missing data, thus I believe that it is 
# A good number of rows to work with 
# and not replace the data
# This also drops all the buildings that are not commercially demolishined
# I think it is very interesting to analyze data of commercial buildings
# If I were to replace the data with mean values, I would write:
# mean_value = df['Column'].mean()
# Filling in the missing values
# df = df.fillna(mean_value)

df.dropna(inplace=True, axis=0)
```

```python
# Here is a list of the data we found in the previous 4 cells
standing = 521
under_construction = 8
demolished = 151
vacant_land = 43

# Calculating the total amount of buildings
total_buildings=(standing+under_construction+demolished+vacant_land)
print(f'The number of buildings in our data set is {total_buildings}.')

# Calculating the percentage of each building status
percent_standing = (standing/total_buildings)*100
print(f'The percentage of buildings still standing is {percent_standing:.2f}%.')
percent_under_construction = (under_construction/total_buildings)*100
print(f'The percentage of buildings under construction is {percent_under_construction:.2f}%.')
percent_demolished = (demolished/total_buildings)*100
print(f'The percentage of buildings demolished is {percent_demolished:.2f}%.')
percent_vacant_land = (vacant_land/total_buildings)*100
print(f'The percentage of buildings turned into vacant land is {percent_vacant_land:.2f}%.')
```

```python
# Data Visualization and Analysis
# Creating a pivot table to show how much of each WorkType is in each Building Status

pivot_table = pd.crosstab(df['BuildingStatus'], df['WorkType'])
print(pivot_table)
```

```python
# Mean and Median of AmountAwarded correlation to the WorkType

result_mean = df.groupby('WorkType')['AmountAwarded'].mean().reset_index()
result_mean.columns = ['WorkType', 'MeanAmountAwarded']

result_median = df.groupby('WorkType')['AmountAwarded'].median().reset_index()
result_median.columns = ['WorkType', 'MedianAmountAwarded']

print("Mean AmountAwarded by WorkType:")
print(result_mean)

print("\nMedian AmountAwarded by WorkType:")
print(result_median)
```

```python
# Mean and Median of AmountAwarded correlation to the Borough

result_mean = df.groupby('Borough')['AmountAwarded'].mean().reset_index()
result_mean.columns = ['Borough', 'MeanAmountAwarded']

result_median = df.groupby('Borough')['AmountAwarded'].median().reset_index()
result_median.columns = ['Borough', 'MedianAmountAwarded']

print("Mean AmountAwarded by Borough:")
print(result_mean)

print("\nMedian AmountAwarded by Borough:")
print(result_median)
```

```python
time_taken =  pd.DataFrame(pd.to_datetime(df['AmountAwardedDate']) - pd.to_datetime(df['ReportCreationDate']))
time_taken['Year'] = pd.to_datetime(df['ReportCreationDate']).dt.year
result_mean = time_taken.groupby('Year').mean() 
result_median = time_taken.groupby('Year').median() 
result_mean.columns = ['MeanTime']
result_median.columns = ['MedianTime']
result_mean = pd.to_timedelta(result_mean['MeanTime']).dt.round('D').reset_index()
result_median = pd.to_timedelta(result_median['MedianTime']).dt.round('D').reset_index()
result_mean.columns = ['Year', 'MeanCompletionTime']
result_median.columns = ['Year', 'MedianCompletionTime']
print(result_mean)
print(result_median)
```

```python
# Calculating how much money is awarded per year

year = pd.to_datetime(df['ReportCreationDate']).dt.year
df = df.assign(Year = year)
result = pd.DataFrame(df.groupby('Year')['AmountAwarded'].sum()).reset_index()
result.columns = ['Year', 'TotalAmountAwarded']
print("Total Amount Awarded per Year:")
print(result)
```

```python
# Calculating how much money is awarded per month
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

date = pd.to_datetime(df['ReportCreationDate'])
df = df.assign(Date = date)
year = pd.to_datetime(df['ReportCreationDate']).dt.year
df = df.assign(Year = year)
month = pd.to_datetime(df['ReportCreationDate']).dt.month
df = df.assign(Month = month)

result = pd.DataFrame(df.groupby(['Year', 'Month'])['AmountAwarded'].sum()).reset_index()
result.columns = ['Year', 'Month', 'TotalAmountAwarded']

print("Total Amount Awarded per Month:")
print(result)
```

```python
# Analyzes the amount awarded mean

result_means = pd.DataFrame(df.groupby('WorkType')['AmountAwarded'].mean()).reset_index()
result_means.columns = ['WorkType', 'TotalAmountAwarded']

print("Mean AmountAwarded by WorkType:")
print(result_means)
```

```python
# Analyze the amount awarded per borough

result_total_amount = df.groupby('Borough')['AmountAwarded'].sum()

result_total_amount = pd.DataFrame(result_total_amount.sort_values(ascending=False)).reset_index()
result_means.columns = ['Borough', 'TotalAmountAwarded']

print("Total Amount Awarded per Borough (Descending Order):")
print(result_total_amount)
```

```python
# Analyzing which type of work invovles more ServiceCharge

result_service_charge = df.groupby('WorkType')['ServiceCharge'].mean()

result_service_charge = pd.DataFrame(result_service_charge.sort_values(ascending=False)).reset_index()
result_service_charge.columns = ['WorkType', 'ServiceCharge']

print("Proportion of ServiceCharge by WorkType (Descending Order):")
print(result_service_charge)
```

```python
# Analyzing which Borough has the most ServiceCharge

result_service_charge = df.groupby('Borough')['ServiceCharge'].mean()

result_service_charge = pd.DataFrame(result_service_charge.sort_values(ascending=False)).reset_index()
result_service_charge.columns = ['Borough', 'ServiceCharge']

print("Proportion of ServiceCharge by Borough (Descending Order):")
print(result_service_charge)
```
